### 23. What is the price of each item? 
- Delete the duplicates in item_name and quantity
- Print a data frame with only two columns `item_name` and `item_price`
- Sort the values from the most to less expensive

```python
# Step 1: Calculate per-unit price per row
chipo['each_item_price'] = chipo['item_price'] / chipo['quantity']

# Step 2: Get one row per item_name (with its unit price)
unique_items = chipo[['item_name', 'each_item_price']].drop_duplicates(subset='item_name')

# Step 3: Sort in descending order of per-unit price
unique_items_sorted = unique_items.sort_values(by='each_item_price', ascending=False, ignore_index=True)

# Output
unique_items_sorted
```

```python
# Step 1: Calculate the total revenue per transaction
online_rt['Revenue'] = online_rt['Quantity'] * online_rt['UnitPrice']

# Step 2: Calculate total revenue by country
revenue_by_country = online_rt.groupby('Country')['Revenue'].sum()

# Step 3: Get the top 3 countries by total revenue
top_3_countries = revenue_by_country.nlargest(3).index

# filter data for top 3 countries
filtered_data = online_rt[online_rt.Country.isin(top_3_countries)]

sns.scatterplot(filtered_data,x='Quantity',y='UnitPrice',hue='CustomerID' )
plt.title("Quantity vs UnitPrice by CustomerID)")
plt.xlabel("Unit Price")
plt.ylabel("Quantity")
plt.show()
```
